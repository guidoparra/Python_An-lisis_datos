```python
import pandas as pd
```

```python
df = pd.read_csv("Archivos csv/nba.csv")
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()  #verifico si existen nulos y sumo la cantidad de cada columna
```

```python
df[df['Age']>df['Age'].mean()].sort_values(by='Age',ascending=False).head() #filtro las edades que sean > al promedio de edad y los ordeno por edad descendente
```

```python
# También podemos usar condiciones de combinación con operadores lógicos como & y |. Por ejemplo, en el siguiente caso, además de la condición anterior, agregamos la 
# condición de que el salario debe ser mayor a 250K. 
# Hay que Tener en cuenta que las condiciones separadas por operadores lógicos deben estar entre paréntesis.

df[(df['Age']>df['Age'].mean()) & (df['Salary'] > 250000)].sort_values(by='Salary',ascending=False).head()
```

```python
df[(df)['Weight']>200.0].sort_values(by='Weight', ascending=True) #jugadores que midan más de 2 metros de altura y ordenados de menor a mayor
```

```python
# LOC accede a los datos por etiqueta (nombres de columnas e índices de filas), mientras que ILOC accede a los datos por índice numérico.
df.loc[(df.Team == 'Los Angeles Lakers') & (df.Age < 23)] 
```

```python
df.iloc[3]
```

```python
df.iloc[0:3] # Recordar que el iloc no trae el último elemento solicitado
```

```python
# A veces necesitamos apuntar a valores específicos. En tales casos, utilizamos ISIN. 
# Por ejemplo, si queremos restringir los valores de la Position

df.loc[df['Position'].isin(['PG','SG']),:].sample(5)
```

```python
# restringir a dos equipos 
condicion = df["Team"].isin([ "Chicago Bulls", "Utah Jazz"]) 

df[condicion].head()
```

```python
condicion2 = df["Team"].str.contains("boston", na = False, case = False)
df[condicion2].head()
```

```python
# Objetivo: Crear una máscara booleana (Series de True/False) que indique qué filas cumplen la condición.
# Detalle de parámetros:
# "boston": Texto a buscar dentro de los valores de la columna "Team".
# na=False: Si hay valores NaN (nulos), los marca como False (no los incluye en el resultado).
# case=False: Ignora mayúsculas/minúsculas ("boston" coincidirá con "Boston", "BOSTON", etc.).

condicion2 = df["Team"].str.contains("boston", na = False, case = False)
df[condicion2].head()
```

```python
# función WHERE
df.where(df["Age"] < 30)
```

```python
# Función QUERY

df.query('Salary < 5000000')
```

```python
# Filtrar las filas del DataFrame df donde los valores de la columna Salary sean menores a 5,000,000.
df.query('Salary < 5000000')
```

```python
df.query("Salary < 5000000 & Age > 25")
```
