```python
import pandas as pd
```

```python
# Técnicas de filtrado utilizando LOC & ILOC
df_books = pd.read_csv('ai_ghibli_trend_dataset_v2.csv', sep=',', header=0)
df_books.loc[:]   #muestra todos los datos del dataFrame
```

```python
df_books.head()
```

```python
# Mostrar un rango de filas tomando en cuenta el start y el end
df_books.loc[0:4]
```

```python
# Filtrando por filas y columnas
df_books.loc[0:4, ['prompt', 'likes']]  #filtra los datos de la fila que va de 0 a 4 y de las columnas Name y Author
```

```python
# Filtrar datos que cumplan una condición determinada
df_books.loc[:, ['platform']] == 'Instagram'
```

```python
# .iloc
# Filtra mediante índices.
df_books.iloc[:] #muestra todos los datos del dataframe
```

```python
# Filtrar datos según los índices de las filas y las columnas
df_books.iloc[:4, 2:5]  # muestra los datos de las filas que van de 0 a 3 y las columnas con indices 0 y 1
```

```python
# Manejo de datos NULOS
# Identificar valores nulos en un DataFrame
df_books.isnull()
```

```python
# Identificar valores nulos con un valor numérico
df_books.isnull()*1
```

```python
# Sustituir los valores nulos por una cadena
df_books.fillna('Sin datos')
```

```python
# Filtrado de datos por condiciones
# Mostrar datos que sean mayores a cierto valor
mayor_tiempo = df_books['generation_time'] > 10.00
print(mayor_tiempo)
```

```python
# También se puede colocar la condición directamente como parámetro
df_books[df_books['generation_time'] > 10.00]   #filtra los datos que cumplen con la condicion
```

```python
# Mostrar los datos que sean igual a cierto valor
prompt = df_books['prompt'] == 'Anime-style train passing through a fantasy world'
prompt #muestra el dataFrame con valores booleanos. True para libros de tipo Fiction
```

```python
df_books.head()
```

```python
# Podemos utilizar la función agg para aplicar diferentes agregaciones a diferentes columnas de un DataFrame.

import seaborn as sns
import pandas as pd

aggregaciones = df_books.agg({
    'likes': ['min', 'max', 'mean'],
    'shares': ['sum', 'mean']
})

agregaciones1 = pd.DataFrame(aggregaciones)

agregaciones1
```

```python
# Podemos utilizar el método isin para filtrar filas según una lista de valores. 
# Este método es útil cuando trabajamos con datos categóricos y deseamos seleccionar filas que pertenecen a determinadas categorías. 

ghibli_filtrado = df_books[df_books['platform'].isin(['Reddit', 'TikTok'])]
ghibli_filtrado.head()
```
